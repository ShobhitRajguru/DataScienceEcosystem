```python
import random
```

```python
a = [random.gauss(50,20) for x in range(30)]
b = [random.gauss(55,15) for x in range(30)]
```

```python
import seaborn as sns
sns.set_style('darkgrid')
sns.kdeplot(a,fill=True)
sns.kdeplot(b,fill=True)
```

```python
import scipy.stats as stats
```

```python
t_stat,p_value=stats.ttest_ind(a,b,equal_var=False)
```

```python
t_stat,p_value
```

```python
import numpy as np
```

```python
np.mean(a),np.mean(b)
```

```python
a = [random.gauss(50,20) for x in range(30)]
b = [random.gauss(60,25) for x in range(30)]
```

```python
t_stat,p_value=stats.ttest_rel(a,b)
```

```python
t_stat,p_value
```

```python
np.mean(b) - np.mean(a)
```

```python
np.mean(a), np.mean(b)
```

```python
t_stat, p_value=stats.ttest_1samp(a,50.001,axis=0)
```

```python
t_stat,p_value
```
